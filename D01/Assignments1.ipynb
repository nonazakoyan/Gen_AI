```python
'''
    Define an abstract base class for a data storage system, 
    with methods like save(), load(), and delete(). 
    Implement concrete subclasses representing different storage systems, 
    such as file-based storage and database storage, ensuring they adhere to the abstract interface.
'''

from abc import ABC, abstractmethod

class Storage(ABC):
    @abstractmethod
    def save(self):
        pass
    @abstractmethod
    def load(self):
        pass
    @abstractmethod
    def delete(self):
        pass

class FileStorage(Storage):
    def save(self):
        print("save the file")
    def load(self):
        print("load the file")
    def delete(self):
        print("delete the file")
        
class DataStorage(Storage):
    def save(self):
        print("save the data")
    def load(self):
        print("load the data")
    def delete(self):
        print("delete the data")
```

```python
'''
    Implement a hierarchy of classes representing different types of vehicles, 
    such as cars, motorcycles, and bicycles. Demonstrate inheritance, method overriding, 
    and polymorphism by implementing common methods and attributes specific to each vehicle type.
'''

class Vehicle:
    def __init__(self, manufacturer, model, year):
        self.manufacturer = manufacturer
        self.model = model
        self.year = year

    def info(self):
        print(f"Manufacturer: {self.manufacturer}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")


class Car(Vehicle):
    def __init__(self, manufacturer, model, year, countofdoors, maxspeed):
        super().__init__(manufacturer, model, year)
        self.countofdoors = countofdoors
        self.maxspeed = maxspeed

    def info(self):
        super().info()
        print(f"Number of doors: {self.countofdoors}")
        print(f"Speed of {self.model} car: {self.maxspeed} km/h")


class Motorcycle(Vehicle):
    def __init__(self, manufacturer, model, year, maxspeed):
        super().__init__(manufacturer, model, year)
        self.maxspeed = maxspeed

    def info(self):
        super().info()
        print(f"Speed of {self.model} motorcycle: {self.maxspeed} km/h")


class Bicycle(Vehicle):
    def __init__(self, manufacturer, model, year, weigh):
        super().__init__(manufacturer, model, year)
        self.weigh = weigh

    def info(self):
        super().info()
        print(f"weigh of {self.model} bicycle: {self.weigh} kg")
```
